```python
def parse_line(line,option):
    if option == 'set':
        return list(set(int(num.strip()) for num in line.strip("{}\n").split(",")))
    return list(int(num.strip()) for num in line.strip("{}\n").split(","))

def converse(string,dictionary):
    res = 0
    for letter in string:
        res = res+(dictionary[letter])
    
    return res

def generate_words(filename,generator_set):
    with open(filename, "r") as file:
        # Read lines from the file
        lines = file.readlines()
    
    # populate the levels array
    levels = [parse_line(line, 'default') for line in lines]
    
    clumped_numerical_level = [parse_line(line, 'set') for line in lines]
    
    # generator to build up words
    generator = generator_set
    
    # quick reference to the letter associated with each generator
    dictionary = {}
    alphabet = [
                'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm',
                'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'
            ]
    for index in range(len(generator)):
        dictionary[str(generator[index])] = alphabet[index]
    reverse_dictionary = {}
    for index in range(len(generator)):
        reverse_dictionary[alphabet[index]] = generator[index]
    print(reverse_dictionary)
    # a 2D array used to store all the distinct words from each level
    stonk = [
        alphabet[:len(generator)]
    ] 
    
    # storage for words at each level
    words = []
    words.append([
        (el,dictionary[str(el)]) for el in generator
    ])
    print(dictionary)
    print()
    
    # give stonk the first stonkers
    
    
    # debug variables
    max_loop = 7
    
    # main loop
    for depth in range(1,len(levels)):
        # for debug only; remove if not needed
        if max_loop == 0:
            break
        max_loop -= 1
            
        # create new words array for the current level
        new_words = []
    
        # loop through element on the current level
        # find the difference between its and the previous elements in the previous level
        for current_element_index in range(len(levels[depth])):
#             print(levels[depth])
            flag = False
            for previous_element_index in range(len(levels[depth-1])):
                diff = levels[depth][current_element_index] - levels[depth-1][previous_element_index]
                if diff in generator:
                    flag = True
                    # loop through every tuple in the previous level to find a match and append the new letter
                    for idx in range(len(words[depth-1])):
                        if words[depth-1][idx][0] == levels[depth-1][previous_element_index]:
                            wrd = words[depth-1][idx][1]+dictionary[str(diff)]
                            new_words.append((levels[depth][current_element_index], wrd))
            if flag == False:
                    print("extra word: {}!\n".format(diff))
        
        words.append(list(set(new_words)))
        baby_stonk = []
        
        # grow baby stonker
        for el in list(set(new_words)):
            baby_stonk.append(el[1])
        
        # feed big stonker
        stonk.append(baby_stonk)
        # remove duplicate words and then display
        print(list(set(new_words)))
        print()
    
    # return value is an array that stores distinct words from each level
    return (stonk, clumped_numerical_level, reverse_dictionary, levels)
```

```python
(papa_stonk, clumped_numerical_level,reverse_dictionary, original_level) = generate_words("simple.txt", [4,5,10,11])
```

```python
def converse(string,dictionary):
    res = 0
    for letter in string:
        res = res+(dictionary[letter])
    
    return res

def clump_words(words, dictionary, clumped_numerical_levels):
    clumped_array = []
    
    # first level
    clumped_array.append(
        [
            (value,key) for key,value in dictionary.items()
        ]
    )
    print(clumped_array)
    
    print(dictionary)
    # populate clumped_array
        
    for level_idx in range(1,len(clumped_numerical_levels)):
        clumped_words = []
        
        for idx in range(len(clumped_numerical_levels[level_idx])):
            current_num = [
                clumped_numerical_levels[level_idx][idx]
            ]
            for word in words[level_idx]:
                if converse(str(word),dictionary) == clumped_numerical_levels[level_idx][idx]:
                    current_num.append(str(word))
                    
            clumped_words.append(current_num)
            
        clumped_array.append(clumped_words)
        
        
    return clumped_array
        
    
arr = clump_words(
    words = papa_stonk,
    dictionary = reverse_dictionary,
    clumped_numerical_levels = clumped_numerical_level[:len(papa_stonk)]
)    
```

```python
import re

pattern = r'(a(b|c|d)*)|((b|c|d)*)'

prune_data = []

for level in arr:
    tmp = []
    for word in level:
        printer = []
        printer.append(word[0])
        for idx in range(1, len(word)):
            if re.fullmatch(pattern, word[idx]):
                printer.append(word[idx])
        print(printer)
        tmp.append(printer)
    prune_data.append(tmp)
```
